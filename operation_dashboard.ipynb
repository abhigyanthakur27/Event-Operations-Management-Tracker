```python
import pandas as pd
import matplotlib.pyplot as plt


# Load data
budget = pd.read_csv('data/budget.csv')


# Plot budget vs actual
plt.figure()
plt.bar(budget['Category'], budget['Allocated'])
plt.bar(budget['Category'], budget['Actual'], bottom=budget['Allocated'])
plt.title('Event Budget Overview')
plt.xlabel('Category')
plt.ylabel('Amount')


# Save dashboard
plt.savefig('dashboards/operations_dashboard.png')
plt.close()
```
